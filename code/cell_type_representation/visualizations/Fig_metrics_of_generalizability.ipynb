### Visualize generalizable cell type embedding space benchmark

The overall score is calculated based on metrics from the paper by [Luecken, et al.](https://www.nature.com/articles/s41592-021-01336-8#Sec10)<br>
Metrics used:
* Isolated Label F1
* CC
* ARI
* NMI
* GC
* Isolated Label ASW
* PCR
* ASW | Batch
* ASW | Bio

```python
files = ["../benchmarks/results/Generalizability/Bone_marrow/Benchmark_results",
         "../benchmarks/results/Generalizability/PBMC/Benchmark_results",
         "../benchmarks/results/Generalizability/Kidney/Benchmark_results",
         "../benchmarks/results/Generalizability/Pancreas/Benchmark_results",
         "../benchmarks/results/Generalizability/All_merged/Benchmark_results"]
dataset_names = ["Bone Marrow Dataset",
                 "PBMC Dataset",
                 "Kidney Dataset",
                 "Pancreas Dataset",
                 "Merged Dataset"]
image_path = "figures/Overall_metrics"
VisEnv.BoxPlotVisualization(files=files, dataset_names=dataset_names, image_path=image_path)
```
