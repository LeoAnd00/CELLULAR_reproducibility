## Preprocessing

```python
# Add QC metrics to adata
adata = dp.QC().QC_metric_calc(adata)
print(adata.obs)
```

```python
qc_adata = dp.QC().QC_filter_outliers(adata,[5,5,5,5])
```

```python
# Investigate which genes are most common in the top 20 genes for cells with a high % counts in the top 20 genes

# Here we can see how the reason for observing cells with high percentage of counts in top 20 genes with highest counts 
# is that these samples mostly consist of hemoglobin related genes.

import numpy as np
import matplotlib.pyplot as plt
import random

top_20_indices = np.argpartition(adata.X, -20, axis=1)[:, -20:]
top_20_values = np.take_along_axis(adata.X, top_20_indices, axis=1)
top_20_frac = (np.sum(top_20_values, axis=1)/adata.obs['n_counts'])

top_20_genes = []
top_20_genes_index = []
for k, frac in enumerate(top_20_frac):
    if frac > 0.8: # Only pick cells with high % counts in the top 20 genes
        for gene_index in top_20_indices[k]:
            top_20_genes.append(adata.var.index[gene_index])
            top_20_genes_index.append(gene_index)

gene_count_dict = {}
for k, gene in enumerate(top_20_genes):
    if gene in gene_count_dict:
        gene_count_dict[gene] += np.sum(adata.X[:,top_20_genes_index[k]][top_20_frac > 0.8])
    else:
        gene_count_dict[gene] = np.sum(adata.X[:,top_20_genes_index[k]][top_20_frac > 0.8])

# Top 20 most common genes in the top 20 genes by counts
sorted_gene_count = sorted(gene_count_dict.items(), key=lambda x: x[1], reverse=True)
common_top_20_genes = dict(sorted_gene_count[:20])

# Barplot
sorted_data = sorted(common_top_20_genes.items(), key=lambda x: x[1], reverse=False)
unique_strings, counts = zip(*sorted_data)

bar_colors = [f'#{random.randint(0, 0xFFFFFF):06x}' for _ in unique_strings]

plt.barh(unique_strings, counts, color=bar_colors)

plt.xlabel('Counts')
plt.ylabel('Gene')
plt.title('Distribution by counts of the most common top 20 genes of cells with above 80% counts in the top 20 genes')

plt.tight_layout()
plt.show()
```
